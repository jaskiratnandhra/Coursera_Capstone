```python
q=data.set_index('SECFNAME').T.to_dict('list')
def code():
    d={}
    from nltk.tokenize import word_tokenize
    from nltk.tokenize import sent_tokenize
    from requests import get
    
    
    for j in q:
        
        f='https://www.sec.gov/Archives/'+j
        url = f
        r = get(url)
        a=r.text
        words = word_tokenize(a)

        sentence=sent_tokenize(a)
        whitelist = set('abcdefghijklmnopqrstuvwxyz ABCDEFGHIJKLMNOPQRSTUVWXYZ')
        qw=[]

        for i in words:
            qw+= [''.join(filter(whitelist.__contains__, i))]
        qwe=[]
        for i in qw:
            if i=='':
                pass
            else:
                qwe+=[i]
        senti=[]
        from textblob import TextBlob
        for i in range(len(qwe)):
            obj=TextBlob(qwe[i])
            sentiment=obj.sentiment.polarity
            senti+=[sentiment]



        p=0 #posistivty rate
        for i in senti:
            if i >0:
                p+=1


        n=0 #negativity rate
        for i in senti:
            if i<0:
                n+=1

        ##Positivity score and negitivity score


        Polarity_Score =((p - n)/((p+n) + 0.000001))



        Subjectivity_Score = (p + n)/ ((len(words)) + 0.000001)
        print(Subjectivity_Score)


        if Polarity_Score<-0.5:
            print("Most Negative")
        elif Polarity_Score>-0.5 and Polarity_Score<0:
            print("Negative")
        elif Polarity_Score==0:
            print("Neutral")
        elif Polarity_Score>0 and Polarity_Score<0.5:
            print("Positive")
        else:
            print("Very Positive")

        Average_Sentence_Length = len(qwe) / len(sentence)
        

        def syllable_count(word):
            word = word.lower()
            count = 0
            vowels = "aeiouy"
            if word[0] in vowels:
                count += 1
            for index in range(1, len(word)):
                if word[index] in vowels and word[index - 1] not in vowels:
                    count += 1
            if word.endswith("e"):
                count -= 1
            if count == 0:
                count += 1
            return count
        sc=[]
        for i in qwe :
            sc+=[syllable_count(i)]
        c=0

        for i in sc:
            if int(i)>1:
                c+=1

        Percentage_of_Complex_words=(c/len(qwe))*100



        Fog_Index = 0.4 * (Average_Sentence_Length +Percentage_of_Complex_words )
        cv=0
        
        for i in qwe:
            if i in ud:
                cv+=1
        cb=0
        cdd=[]
        for i in qwe:
            if i in cd:
                cb+=1
                cdd+=[i]
        s=q[j]
        
        pwp=p/len(qwe)
        nwp=n/len(qwe)
        uwp=cv/len(qwe)
        cwp=cb/len(qwe)
        
        s.append(p)
        s.append(n)
        s.append(Polarity_Score)
        s.append(Average_Sentence_Length)
        s.append(Percentage_of_Complex_words)
        s.append(Fog_Index)
        s.append(c)
        s.append(len(qwe))
        s.append(cv)
        s.append(cb)
        s.append(pwp)
        s.append(nwp)
        s.append(uwp)
        s.append(cwp)
        s.append(cdd)
        
    print(q)
code()
```
